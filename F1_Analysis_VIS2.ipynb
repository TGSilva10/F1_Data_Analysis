```python
import pandas as pd
import matplotlib.pyplot as plt

# Função para visualizar resultados de queries fictícias
def visualize_results(data, title, x_label, y_label, legend_label=None):
    plt.figure(figsize=(10, 6))
    for key, group in data.groupby(legend_label):
        plt.plot(group[x_label], group[y_label], label=key, marker='o')
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    if legend_label:
        plt.legend(loc='best')
    plt.show()

# Dados fictícios para demonstração

# 1. Campeão com Maior Número de Poles
data_poles = pd.DataFrame({
    'driver_name': ['Hamilton', 'Schumacher', 'Senna', 'Vettel', 'Prost'],
    'pole_positions': [68, 55, 65, 57, 33]
})

# Gráfico de barras para número de poles
plt.figure(figsize=(10, 6))
plt.bar(data_poles['driver_name'], data_poles['pole_positions'], color='skyblue')
plt.title('Número de Poles Positions por Piloto')
plt.xlabel('Piloto')
plt.ylabel('Número de Poles')
plt.show()
```

```python
# 2. Comparação de Construtores por Décadas
data_constructors_decades = pd.DataFrame({
    'constructor_name': ['Mercedes', 'Ferrari', 'McLaren', 'Red Bull', 'Williams'],
    'decade': [2010, 2010, 2000, 2020, 1990],
    'wins': [100, 45, 75, 40, 60]
})

# Gráfico de Linhas para construtores por décadas
visualize_results(data_constructors_decades, 'Vitórias de Construtores por Década', 'decade', 'wins', 'constructor_name')
```

```python
# 3. Melhor Desempenho em Circuitos Europeus vs Não-Europeus
data_regions = pd.DataFrame({
    'region': ['Europe', 'Non-Europe'],
    'wins': [400, 300]
})

# Gráfico de Pizza para vitórias por região
plt.figure(figsize=(8, 8))
plt.pie(data_regions['wins'], labels=data_regions['region'], autopct='%1.1f%%', startangle=90, colors=['royalblue', 'lightcoral'])
plt.title('Vitórias em Circuitos Europeus vs Não-Europeus')
plt.show()
```

```python
# 4. Análise de Desempenho ao Longo do Tempo para Pilotos
data_pilot_performance = pd.DataFrame({
    'driver_name': ['Hamilton', 'Hamilton', 'Schumacher', 'Schumacher'],
    'year': [2017, 2018, 2001, 2002],
    'wins': [10, 11, 9, 11]
})

# Visualização de desempenho ao longo do tempo
visualize_results(data_pilot_performance, 'Desempenho de Pilotos ao Longo do Tempo', 'year', 'wins', 'driver_name')
```

```python
# 5. Análise de Construtores ao Longo do Tempo
data_constructor_performance = pd.DataFrame({
    'constructor_name': ['Mercedes', 'Mercedes', 'Ferrari', 'Ferrari'],
    'year': [2017, 2018, 2001, 2002],
    'wins': [12, 13, 10, 11]
})

# Visualização de desempenho de construtores ao longo do tempo
visualize_results(data_constructor_performance, 'Desempenho de Construtores ao Longo do Tempo', 'year', 'wins', 'constructor_name')
```
